# First dataset 

```python
religion = pd.read_csv('world_religions.csv')
```

```python
asia_religion = religion[religion['region'] == 'Asia']
asia_religion.columns
```

```python
asia_religion = asia_religion[['year','christianity_all','judaism_all','islam_all','buddhism_all','zoroastrianism_all',
                              'hinduism_all', 'sikhism_all', 'shinto_all','baha’i_all', 'taoism_all', 'jainism_all', 
                              'confucianism_all','syncretism_all', 'animism_all', 'noreligion_all', 'otherreligion_all']]
```

```python
asia_religion['All_other_religions'] = asia_religion['judaism_all']+asia_religion['zoroastrianism_all']+asia_religion ['shinto_all'] + asia_religion ['baha’i_all'] + asia_religion ['taoism_all']+asia_religion ['jainism_all']+asia_religion ['confucianism_all']+asia_religion ['syncretism_all']+asia_religion ['animism_all']+asia_religion ['noreligion_all']+asia_religion['sikhism_all']+asia_religion['buddhism_all']+asia_religion['otherreligion_all']
asia_religion .drop(['buddhism_all','judaism_all','zoroastrianism_all','sikhism_all','shinto_all','baha’i_all', 'taoism_all',
                     'jainism_all', 'confucianism_all','syncretism_all', 'animism_all', 'noreligion_all',
                     'otherreligion_all'],axis = 1,inplace = True)
```

```python
asia_religion
```

# Second Dataset

```python
years_array = asia_religion['year'].to_numpy()
christianity_array = asia_religion['christianity_all'].to_numpy()/10**8
islam_array = asia_religion['islam_all'].to_numpy()/10**8
hinduism_array = asia_religion['hinduism_all'].to_numpy()/10**8
All_other_religions_array = asia_religion['All_other_religions'].to_numpy()/10**8
```

```python
world_population = pd.read_csv('world_regions.csv')
countries = world_population.loc[world_population['Year'] == '1945']
countries['Entity'].to_numpy()
```

```python
world_population = world_population.loc[world_population['Year'].isin(['1945','1950','1955','1960','1965','1970','1975',
                                                                       '1980','1985','1990','1995','2000','2005','2010'])]
world_population.reset_index(drop = True,inplace = True)
```

```python
asia_population = world_population.loc[world_population['Entity'].isin(['Afghanistan','Armenia','Azerbaijan','Bahrain',
                                                                        'Bangladesh','Bhutan','Brunei','Cambodia','China',
                                                                       'Egypt','Georgia','Hong Kong','India','Indonesia',
                                                                        'Iran','Iraq','Israel','Japan', 'Jordan','Kazakhstan',
                                                                        'Kuwait', 'Kyrgyzstan', 'Laos','Lebanon', 'Malaysia',
                                                                        'Maldives','Mongolia','Myanmar','Nepal','North Korea',
                                                                        'Oman', 'Pakistan','Palestine','Papua New Guinea',
                                                                        'Philippines','Qatar','Russia','Saudi Arabia',
                                                                        'Singapore','South Korea','Sri Lanka','Taiwan',
                                                                        'Tajikistan','Thailand','Turkey', 'Turkmenistan',
                                                                        'United Arab Emirates','Uzbekistan','Vietnam','Yemen'])]
```

```python
b = asia_population
b = b.drop(['Entity','Code'],axis = 1)
asiapopulation_array = b.groupby(['Year']).sum().to_numpy()
asiapopulation_array =asiapopulation_array/ 10**8 
```

# Plotting Data

```python
from pylab import rcParams
rcParams['figure.figsize'] = 10,6

plt.figure()

plt.suptitle(t= 'Growth of Various religions vs Population in Asia',fontsize=19,fontfamily = 'Times New Roman')
plt.xlabel('Year',labelpad = 15,size=12)
plt.ylabel('Population in multiples of 10^8',labelpad = 15,size=12)

plt.plot(years_array,asiapopulation_array, '-o',label = 'Asia Population')
plt.plot(years_array, christianity_array, '-.s',label = 'Christian')
plt.plot(years_array,islam_array,'-.^',label = 'Islam')
plt.plot(years_array,hinduism_array,'-.p',label = 'Hindu')
plt.plot(years_array,All_other_religions_array,'-.kH',label = 'Other Religions')

plt.xticks(np.arange(min(years_array), max(years_array)+1, 5.0))
plt.yticks(np.arange(0,51, 5.0))


w = plt.gca().xaxis

for item in w.get_ticklabels():
    item.set_rotation(45)
    
plt.subplots_adjust(bottom = 0.18)

for d in ["top", "right"]:
    plt.gca().spines[d].set_visible(False)


plt.legend(loc = 'best',fancybox = True,edgecolor = '#F5F5F5', title = 'Legend',title_fontsize = 11,facecolor='#F5F5F5')
```
